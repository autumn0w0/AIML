WAF to print the length of a list. (list is the parameter)


```python
def len_list(lst):
    return len(lst)

print(len_list([1, 2, 3, 4, 5]))
```

WAF to print the elements of a list in a single line. (list is the parameter)


```python
lst = [1, 2, 3, 4, 5]

def print_list(lst):
    for item in lst:
        print(item, end=' ')

print_list(lst)
```

WAF to find the factorial of n. (n is the parameter)


```python
def cal_factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print(cal_factorial(6))
```

WAF to convert USD to INR.


```python
def con_usd_to_inr(usd):
    return usd * 93.32

print(con_usd_to_inr(100))
```

WAF that will restrn "odd" if the input number is odd else "even"


```python
n = int(input("Enter a number: "))

def odd_or_even(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(odd_or_even(n))
```
